###### Creating some more data - list with more positive and negative integers

```python
a = [0, -100, 25, 67, -323]
b = [0,3,7,3,9]
fig2 = plt.plot(a,b)
fig2
```

Say we want to focus on just the triangle portion of the plot. To do that, we can ask matplotlib to focus on the specific coordinates to focus on. The x-axis for the left vertex of the triange is around -50, and the x-axis for the right vertex of the triangle is around 70. The y axis for the right vertex of the triangle is around 3 and the y axis for top vertex of the triangle is around 7.

```python
plt.axis([-50, 70, 2.5, 7.5])
fig3 = plt.plot(a,b)
fig3
```

We could set the title for the plot as well as set labels for both axes using the following commands. We could also rename the ticks along x or y axis as shown. 

```python
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.axis([-50, 80, 2, 8])
plt.xticks((-40, -20, 0, 20, 40, 60, 80), ('h', 'e', 'l', 'l', 'o', 'o', 'o'))
fig4 = plt.plot(a,b)
fig4
```

###### Histograms

Histograms allow us to create bins for our data and visualize how many values fall into those bins. These plots help us to understand the distribution of the data. 

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
```

We can specify the number of bins we would like in our histogram as well as the edge color so that we can differentiate clearly between bins in the plot. 
We can also set values of custom bins. 

```python
plt.title("Age of Students")
plt.xlabel("Age")
plt.ylabel("Total Students")
bins = [10,20,30,40,50,60]
plt.hist(ages, bins = bins, edgecolor='white')
```

we could also remove a bin from the bins list if we wanted to explude that data from the plot. In the example, below, no ages between 10-20 are shown. 

```python
plt.title("Age of Students")
plt.xlabel("Age")
plt.ylabel("Total Students")
bins = [20,30,40,50,60]
plt.hist(ages, bins = bins, edgecolor='white')
```
